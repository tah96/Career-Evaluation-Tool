```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
area_code_table = "area-titles.csv"
area_table = pd.read_csv(area_code_table)
area_table.head()
```

```python
area_table = area_table.rename(columns = {"area_fips": "Area Code", "area_title": "Area Name"})
area_table.head()
```

```python
area_table_clean = area_table.set_index(keys="Area Code")
area_table_clean.head()
```

```python
area_table_clean.to_csv("cleaned_data/area_codes.csv", header=True)
```

```python
keywords = ["TOTAL","combined", "Statewide", "Puerto Rico", "Virgin Islands", "Unknown", "Multicounty", "Out-of-State", "Overseas", "MicroSA", "MSA", "CSA", "Athens-Clarke"]

searchstring = '|'.join(keywords)

county_data = area_table[area_table['Area Name'].str.contains(searchstring)==False]
county_data.head()
```

```python
keywords2 = ["County", "Census Area", "Borough"]

searchstring2 = '|'.join(keywords2)

new_county_data = county_data[county_data['Area Name'].str.contains(searchstring2)]
new_county_data.head()
```

```python
county_data_clean = new_county_data.set_index(keys="Area Code")
county_data_clean.head()
```

```python
county_data_clean.to_csv("cleaned_data/county_codes.csv", header=True)
```

```python
keywords3 = ["TOTAL","combined", "Statewide"]

searchstring3 = '|'.join(keywords3)

state_data = area_table[area_table['Area Name'].str.contains(searchstring3)]
state_data.head()
```

```python
keywords4 = ["Multicounty", "Multi", "Puerto Rico", "Virgin Islands", "combined", "TOTAL"]

searchstring4 = '|'.join(keywords4)

new_state_data = state_data[state_data['Area Name'].str.contains(searchstring4)==False]
new_state_data.head()
```

```python
state_data_clean = new_state_data.set_index(keys="Area Code")
state_data_clean.head()
```

```python
state_data_clean.to_csv("cleaned_data/state_codes.csv", header=True)
```

```python
national_geo_data = "national_2019_data.csv"
national_table = pd.read_csv(national_geo_data)
national_table.head()
```

```python
codes = ['15-', '23-', '25-', '29-',
           '31-', '41-']

searchstr = '|'.join(codes)
national_table1 = national_table[national_table['occ_code'].str.contains(searchstr)]
national_table1
```

```python
detailed = ['detailed']

searchstr6 = '|'.join(detailed)
national_table2 = national_table1[national_table1['o_group'].str.contains(searchstr6)]
national_table2
```

```python
national_table3 = national_table2[[ "area", "occ_code", "occ_title","own_code", "tot_emp", "h_mean", "h_pct10", "h_pct25", "h_median", "h_pct75", "h_pct90", "a_mean", "a_pct10", "a_pct25", "a_median", "a_pct75", "a_pct90"]]
national_table3
```

```python
national_data_clean = national_table3.rename(columns = {"area": "Area Code", "occ_code": "Occupation Code", "occ_title": "Occupation Title","own_code": "Ownership Code", "tot_emp": "Total Employment", "h_mean": "Mean Hourly Income", "h_pct10": "Hourly 10th Percentile", "h_pct25": "Hourly 25th Percentile", "h_median": "Median Hourly Income", "h_pct75": "Hourly 75th Percentile", "h_pct90": "Hourly 90th Percentile", "a_mean": "Mean Annual Income", "a_pct10": "Annual 10th Percentile", "a_pct25": "Annual 25th Percentile", "a_median": "Median Annual Income", "a_pct75": "Annual 75th Percentile", "a_pct90": "Annual 90th Percentile"})
national_data_clean
```

```python
national_data_clean.to_csv("cleaned_data/national_emp_data.csv", header=True)
```
