```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
area_code_table = "area-titles.csv"
area_table = pd.read_csv(area_code_table)
area_table
```

```python
area_table = area_table.rename(columns = {"area_fips": "Area Code", "area_title": "Area Name"})
area_table
```

```python
area_table_clean = area_table.set_index(keys="Area Code")
area_table_clean
```

```python
area_table_clean.to_csv("cleaned_data/area_codes.csv", header=True)
```

```python
industry_code_table = "industry-titles.csv"
industry_table = pd.read_csv(industry_code_table)
industry_table
```

```python
new_industry_table = industry_table.rename(columns = {"industry_code": "Industry Code", "industry_title": "Industry Name"})
new_industry_table
```

```python
industry_table_clean = new_industry_table.set_index(keys="Industry Code")
industry_table_clean
```

```python
industry_table_clean.to_csv("cleaned_data/industry_codes.csv", header=True)
```
