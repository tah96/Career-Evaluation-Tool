```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
area_code_table = "area-titles.csv"
area_table = pd.read_csv(area_code_table)
area_table
```

```python
area_table = area_table.rename(columns = {"area_fips": "Area Code", "area_title": "Area Name"})
area_table
```

```python
area_table_clean = area_table.set_index(keys="Area Code")
area_table_clean
```

```python
area_table_clean.to_csv("cleaned_data/area_codes.csv", header=True)
```

```python
keywords = ["TOTAL","combined", "Statewide", "Puerto Rico", "Virgin Islands", "Unknown", "Multicounty", "Out-of-State", "Overseas", "MicroSA", "MSA", "CSA", "Athens-Clarke"]

searchstring = '|'.join(keywords)

county_data = area_table[area_table['Area Name'].str.contains(searchstring)==False]
county_data
```

```python
keywords2 = ["County", "Census Area", "Borough"]

searchstring2 = '|'.join(keywords2)

new_county_data = county_data[county_data['Area Name'].str.contains(searchstring2)]
new_county_data
```

```python
county_data_clean = new_county_data.set_index(keys="Area Code")
county_data_clean
```

```python
county_data_clean.to_csv("cleaned_data/county_codes.csv", header=True)
```

```python
keywords3 = ["TOTAL","combined", "Statewide"]

searchstring3 = '|'.join(keywords3)

state_data = area_table[area_table['Area Name'].str.contains(searchstring3)]
state_data
```

```python
keywords4 = ["Multicounty", "Multi", "Puerto Rico", "Virgin Islands", "combined", "TOTAL"]

searchstring4 = '|'.join(keywords4)

new_state_data = state_data[state_data['Area Name'].str.contains(searchstring4)==False]
new_state_data.head()
```

```python
state_data_clean = new_state_data.set_index(keys="Area Code")
state_data_clean
```

```python
state_data_clean.to_csv("cleaned_data/state_codes.csv", header=True)
```
