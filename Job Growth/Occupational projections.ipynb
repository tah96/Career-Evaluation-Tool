```python
import pandas as pd
import numpy as np
```

```python
occupational_proj_df = pd.read_csv("Occupational projections.csv", usecols=range(0,9)).dropna()
occupational_proj_df.head(10)
```

```python
occupational_proj_df = occupational_proj_df.rename(columns=str).rename(columns={
    "Table 1.7 Occupational projections, 2019-29, and worker characteristics, 2019\n (Numbers in thousands)":"Title",
    "Unnamed: 1": "Code",
    "Unnamed: 2": "Type",
    "Unnamed: 3": "Employment_A",
    "Unnamed: 4": "Employment_B",
    "Unnamed: 5": "Ten_Year_Change",
    "Unnamed: 6": "Ten_Year_Percent_Change",
    "Unnamed: 7": "Percent_self_employed",
    "Unnamed: 8":"Occupational_openings_Ten_Year_Annual_Average",
})
occupational_proj_df
```

```python
occupational_proj_df["Occupation Type"]
```

```python
growth_df = occupational_proj_df.loc[occupational_proj_df["Occupation Type"] == "Line item"]
growth_df
```

```python
growth_df["Employment Matrix code"].values[0][0:2]
```

```python
codes = ['15-', '23-', '25-', '29-',
           '31-', '41-']
searchstr = '|'.join(codes)
occ_growth = growth_df[growth_df['Employment Matrix code'].str.contains(searchstr)] 
occ_growth
```

```python
occ_growth.to_csv("projected_growth.csv")
```
