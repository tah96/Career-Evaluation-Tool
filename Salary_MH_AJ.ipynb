```python
import pandas as pd 
import numpy as np
```

```python
occupations_df = pd.read_csv("Occupations.csv", encoding='latin-1', header=[1]).dropna()
occupations_df.head(10)
```

```python
occupations_df = occupations_df.rename(columns={ "2019 National Employment Matrix title and code":"Title",
                                                          "Unnamed: 1":"Code",
                                                          "Occupation\ntype":"Occupation_Type",
                                                          "Employment":"Employment_2019",
                                                          "Unnamed: 4":"Employment_2029",
                                                          "Employment change, 2019-29":"Employment_Changes_2019_29",
                                                          "Unnamed: 6":"Percent_Employment_Change_2019_29",
                                                          "Percent self employed, 2019":"Percent_Self_Employed_2019",
                                                          "Occupational openings, 2019-29 annual average":"Average_Openings_2019-29",
                                                          "Median annual wage, 2019(1)":"Median_Annual_Wage",
                                                         })
list(occupations_df.columns)
```

```python
occupations_df.columns
```

```python
occupations1b_df = occupations_df[occupations_df["Code"].str.match("15-*") == True]
occupations1c_df = occupations_df[occupations_df["Code"].str.match("23-*") == True]
occupations1d_df = occupations_df[occupations_df["Code"].str.match("25-*") == True]
occupations1e_df = occupations_df[occupations_df["Code"].str.match("29-*") == True]
occupations1f_df = occupations_df[occupations_df["Code"].str.match("31-*") == True]
occupations1g_df = occupations_df[occupations_df["Code"].str.match("41-*") == True]
SalaryFINAL_df = occupations1b_df.append(occupations1c_df).append(occupations1d_df).append(occupations1e_df).append(occupations1f_df).append(occupations1g_df)
SalaryFINAL_df
```

```python
SalaryFINAL_df = SalaryFINAL_df.dropna()
SalaryFINAL_df = SalaryFINAL_df[['Title', 'Code', 'Employment_2019','Employment_2029', 'Employment_Changes_2019_29','Percent_Employment_Change_2019_29', 'Percent_Self_Employed_2019',
       'Average_Openings_2019-29', 'Median_Annual_Wage']]
```

```python
SalaryFINAL_df.to_csv("SalaryFINAL.csv")
```
