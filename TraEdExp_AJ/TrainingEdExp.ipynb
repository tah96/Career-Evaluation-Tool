```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import re
```

### Store CSV into DataFrame

```python
csv1_file = "Resources/Table_1.7.csv"
TrainingEdExp1_df = pd.read_csv(csv1_file, header=[1])
TrainingEdExp1_df.head(5)
```

```python
list(TrainingEdExp1_df.columns)
```

```python
TrainingEdExp1_df = TrainingEdExp1_df.rename(columns = {
    "2019 National Employment Matrix title and code" : "Employment Matrix Title",
    "Unnamed: 1" : "Employment Matrix Code"
})

list(TrainingEdExp1_df.columns)
```

```python
TrainingEdExp1a_df = TrainingEdExp1_df[['Employment Matrix Title','Employment Matrix Code','Typical education needed for entry',
 'Work experience in a related occupation','Typical on-the-job training needed to attain competency in the occupation']]
TrainingEdExp1a_df.head(5)
TrainingEdExp1a_df.dtypes
```

```python
TrainingEdExp1b_df = TrainingEdExp1a_df[TrainingEdExp1a_df["Employment Matrix Code"].str.match("15-*") == True]
TrainingEdExp1c_df = TrainingEdExp1a_df[TrainingEdExp1a_df["Employment Matrix Code"].str.match("23-*") == True]
TrainingEdExp1d_df = TrainingEdExp1a_df[TrainingEdExp1a_df["Employment Matrix Code"].str.match("25-*") == True]
TrainingEdExp1e_df = TrainingEdExp1a_df[TrainingEdExp1a_df["Employment Matrix Code"].str.match("29-*") == True]
TrainingEdExp1f_df = TrainingEdExp1a_df[TrainingEdExp1a_df["Employment Matrix Code"].str.match("31-*") == True]
TrainingEdExp1g_df = TrainingEdExp1a_df[TrainingEdExp1a_df["Employment Matrix Code"].str.match("41-*") == True]
TrainingEdExp1FINAL_df = TrainingEdExp1b_df.append(TrainingEdExp1c_df).append(TrainingEdExp1d_df).append(TrainingEdExp1e_df).append(TrainingEdExp1f_df).append(TrainingEdExp1g_df)
TrainingEdExp1FINAL_df["Employment Matrix Code"]
TrainingEdExp1FINAL_df = TrainingEdExp1FINAL_df.rename(columns = {'Work experience in a related occupation' : "Related Work Experience",
'Typical on-the-job training needed to attain competency in the occupation' : 'Typical on-the-job training'})
TrainingEdExp1FINAL_df.head(5)
TrainingEdExp1FINAL_df.dtypes
TrainingEdExp1FINAL_df.to_csv("TrainingEdExp1FINAL.csv")
```

```python
csv2_file = "Resources/Table_5.3.csv"
TrainingEdExp2_df = pd.read_csv(csv2_file, header=[1])
TrainingEdExp2_df.head(5)
```

```python
list(TrainingEdExp2_df.columns)
```

```python
TrainingEdExp2_df = TrainingEdExp2_df.rename(columns = {
    "2019 National Employment Matrix title and code" : "Employment Matrix Title",
    "Unnamed: 1" : "Employment Matrix Code",
})

list(TrainingEdExp2_df.columns)

TrainingEdExp2_df = TrainingEdExp2_df[['Employment Matrix Code','Less than high school diploma','High school diploma or equivalent','Some college, no degree',"Associate's degree","Bachelor's degree","Master's degree",'Doctoral or professional degree']]
TrainingEdExp2_df.head(5)
```

```python
TrainingEdExp2b_df = TrainingEdExp2_df[TrainingEdExp2_df["Employment Matrix Code"].str.match("15-*") == True]
TrainingEdExp2c_df = TrainingEdExp2_df[TrainingEdExp2_df["Employment Matrix Code"].str.match("23-*") == True]
TrainingEdExp2d_df = TrainingEdExp2_df[TrainingEdExp2_df["Employment Matrix Code"].str.match("25-*") == True]
TrainingEdExp2e_df = TrainingEdExp2_df[TrainingEdExp2_df["Employment Matrix Code"].str.match("29-*") == True]
TrainingEdExp2f_df = TrainingEdExp2_df[TrainingEdExp2_df["Employment Matrix Code"].str.match("31-*") == True]
TrainingEdExp2g_df = TrainingEdExp2_df[TrainingEdExp2_df["Employment Matrix Code"].str.match("41-*") == True]
TrainingEdExp2FINAL_df = TrainingEdExp2b_df.append(TrainingEdExp2c_df).append(TrainingEdExp2d_df).append(TrainingEdExp2e_df).append(TrainingEdExp2f_df).append(TrainingEdExp2g_df)
TrainingEdExp2FINAL_df["Employment Matrix Code"]
TrainingEdExp2FINAL_df.dtypes
TrainingEdExp2FINAL_df.to_csv("TrainingEdExp2FINAL.csv")
```

```python
#Merge TrainingEdExp1FINAL_df & TrainingEdExp2FINAL_df.  
TraEdExp_FINAL_df = pd.merge(TrainingEdExp1FINAL_df, TrainingEdExp2FINAL_df, on="Employment Matrix Code", how="outer")
TraEdExp_FINAL_df
```

```python
#Reorder columns and drop all rows with nulls.  
list(TraEdExp_FINAL_df.columns)

TraEdExp_FINAL_df = TraEdExp_FINAL_df[['Employment Matrix Title','Employment Matrix Code','Typical education needed for entry','Less than high school diploma','High school diploma or equivalent','Some college, no degree',"Associate's degree","Bachelor's degree","Master's degree",'Doctoral or professional degree', 'Related Work Experience','Typical on-the-job training']]

TraEdExp_FINAL_df.head(5) 

TraEdExp_FINAL_df = TraEdExp_FINAL_df.dropna()

TraEdExp_FINAL_df
```
